```python
# mustafa
# pie chart
import matplotlib.pyplot as plt
```

```python
categories = ['pants' , 'shirts' , 'outerwear' , 'pajamas' , 'underweear' , 'shoes']
revenue = [25, 15, 5, 20, 10,25]
```

```python
plt.pie(revenue, labels = categories, autopct='%1.1f%%')
plt.show()
```

```python
explode = [0,.1,0,0,0,0] 
plt.pie(revenue, labels = categories, autopct='%.1f%%', explode = explode)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
categories = ['pants' , 'shirts' , 'outerwear' , 'pajamas' , 'underweear' , 'shoes']
revenue = [25, 15, 5, 20, 10,25]
explode_values= [0,.1,0,0,0,0] 
plt.pie(revenue, labels = categories, autopct='%1.1f%%', explode = explode_values)
plt.show()
```

```python
# Assign 6 colors
colors = ["yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8, 8))
categories = ['pants' , 'shirts' , 'outerwear' , 'pajamas' , 'underweear' , 'shoes']
revenue = [25, 15, 5, 20, 10,25]
explode_values= [0,.4,0,0,0,0] 
plt.pie(revenue, labels = categories, autopct='%.1f%%', explode = explode_values, colors=colors)
plt.show()
```

```python
# object-oriented approach
#plt.subplots(figsize=(8, 8))
#fig= plt.figure(figsize=(6,3))
fig, ax = plt.subplots()
ax.pie(revenue,labels=categories, autopct='%.1f%%',explode = explode_values, colors=colors, shadow="true")

fig = plt.gcf()
fig.set_size_inches(8,8) # or (4,4) or (5,5) or whateverplt.show() https://stackoverflow.com/questions/29786807/how-to-make-a-pie-chart-smaller-with-matplotlib
plt.show()
```
